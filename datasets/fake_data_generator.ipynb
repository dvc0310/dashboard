```python
import pandas as pd
import numpy as np
from faker import Faker
import random

# Initialize Faker to generate fake data
fake = Faker()

# Constants
n_rows = 200000
companies = [
    "VERIZON WIRELESS", "COMCAST BUSINESS COMMUNICATIONS, LLC", "GCI", "AT&T INC", 
    "TIME WARNER CABLE", "COMCAST IP PHONE, LLC", "COMCAST PHONE, LLC", 
    "LEVEL 3 COMMUNICATIONS, LLC", "CHARTER", "T-MOBILE", "XO COMMUNICATIONS", 
    "US CELLULAR", "VERIZON BUSINESS", "LUMEN", "AT&T Mobility", "VERIZON", 
    "TDS TELECOM", "QCC - QWEST COMMUNICATIONS CORP.", "FRONTIER COMMUNICATIONS"
]
statuses = ["Final", "InProgress", "Pending"]
cities = ["Gainesville", "San Francisco", "New York", "Chicago"]
yes_no = ["Yes", "No"]

# Data generation
data = {
    "u_legacy_smartcode_outage_id": [f"{random.randint(1, 99)}-{random.randint(100000, 999999)}" for _ in range(n_rows)],
    "u_outage_report_status": [random.choice(statuses) for _ in range(n_rows)],
    "u_company": [random.choice(companies) for _ in range(n_rows)],
    "u_previous_report_status": [random.choice(statuses) for _ in range(n_rows)],
    "u_incident_date_time": [fake.date_between(start_date="-8y", end_date="today") for _ in range(n_rows)],
    "u_duration_hours": np.random.randint(1, 25, n_rows),
    "u_duration_minutes": np.random.randint(0, 60, n_rows),
    "u_states_affected": ["" for _ in range(n_rows)],
    "u_city_affected": [random.choice(cities) for _ in range(n_rows)],
    "u_incident_description": ["Random incident description" for _ in range(n_rows)],
    "u_cause_description": ["" for _ in range(n_rows)],
    "u_wireline_affected_ind": [random.choice(yes_no) for _ in range(n_rows)],
    "u_wireless_affected_ind": [random.choice(yes_no) for _ in range(n_rows)],
    "u_voip_affected_ind": [random.choice(yes_no) for _ in range(n_rows)],
    "u_number_of_blocked_calls": np.random.randint(0, 1000, n_rows),
    "u_ds3s": np.random.randint(0, 10, n_rows),
    "u_assignment.u_nors_assign_reason1": ["" for _ in range(n_rows)],
    "u_assignment.u_nors_assign_person1": ["" for _ in range(n_rows)],
    "u_assignment.u_nors_assign_reason2": ["" for _ in range(n_rows)],
    "u_assignment.u_nors_assign_person2": ["" for _ in range(n_rows)],
    "u_assignment.u_nors_assign_groupnum": ["" for _ in range(n_rows)]
}

# Convert dictionary to DataFrame
df_fake_outages = pd.DataFrame(data)

# Output sample and info
df_fake_outages.info(), df_fake_outages.head()
```
